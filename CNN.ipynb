```python
import tensorflow as tf

from tensorflow.keras import models,layers,datasets

import matplotlib.pyplot as plt

import numpy as np
```

```python
(x_train, y_train),(x_test, y_test)=datasets.mnist.load_data()

x_train_ann=x_train/255

x_test_ann=x_test/255

plt.imshow(x_train[0])

x_train.shape
```

```python
cnn=models.Sequential([

layers.Conv2D(filters=28, kernel_size=(3,3), activation="relu",input_shape=(28,28,1)),

layers.MaxPooling2D((2,2)),

layers.Conv2D(filters=28, kernel_size=(3,3), activation="relu"),

layers.MaxPooling2D((2,2)),

layers.Flatten(),

layers.Dense(100,activation="relu"),

layers.Dense(10,activation="softmax")

])

cnn.compile(optimizer='adam',

loss='sparse_categorical_crossentropy',

metrics=['accuracy'])

cnn.fit(x_train_ann,y_train,epochs=5)

pred=cnn.predict(x_test_ann)
```

```python
for i in range(10):
    
    plt.subplot(2,5,i+1)

    plt.imshow(x_test[i])

    plt.xlabel(np.argmax(pred[i]))
```
